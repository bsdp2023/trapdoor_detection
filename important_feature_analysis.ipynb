```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from experiment_utils import *

importance_features = pd.read_csv("lightgbm_important_features.csv")
sums = importance_features.mean(axis=0, skipna=True)
sums = sums[sums > 0]
sums = sums.sort_values(ascending=False)
display(sums.head(10))
all_features_list = sums.index.values
top10 = sums.head(10).reset_index()
top10.columns = ['Features', 'Importance']

plt.figure(figsize=(10, 3))
sns.set_color_codes("pastel")
sns.barplot(x = 'Importance', y = 'Features', data = top10)
plt.title('The Importance of the Ten Most Significant Features')
plt.tight_layout()
plt.show()
```
